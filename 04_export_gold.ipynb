```python
EXPORT_PATH = "/Volumes/shopflow/default/shopflow_row/gold_export"

gold_tables = [
 "daily_revenue",
 "monthly_revenue",
 "top_products",
 "category_performance",
 "customer_segments"
]

for tbl in gold_tables:
    df = spark.table(f"shopflow.gold.{tbl}")
    df.coalesce(1) \
        .write \
            .mode("overwrite") \
                .option("header", "true") \
                    .csv(f"{EXPORT_PATH}/{tbl}")
    print(f"Exported {tbl}")

print("\n All gold table exported")
```
