```python
from faker import Faker
import pandas as pd
import random

fake = Faker()
random.seed(42)
Faker.seed(42)
```

```python
CATEGORES = [
    "Electornic", "Clothing", "Home and Kichen", "Sports", "Book", "Beauty", "Toys"
]

PRODUCT_NAME={
    "Electornic": ["wirelees earbuds", "usb-c hub","phone stad"],
    "Clothing": ["Running shoes","Hoodie","Denum Jacket"],
    "Home and Kichen": ["Air fyer","coffee maker", "Blender"],
    "Sports": ["Yoga Mat","Water Bottle","Resistance Bands"],
    "Book": ["Python cookbook", "Data Engineering Book "],
    "Beauty": ["Face Serum","Sunscreen SPF50","Lip Balm"],
    "Toys": ["LEGO Set","Board Game","Puzzle 1000pc"]
}

STATUSES = ["completed","completed","completed",
            "pending","returned"]

CATEGORES = [
    "Electornic", "Clothing", "Home and Kichen", "Sports", "Book", "Beauty", "Toys"
]

PRODUCT_NAME={
    "Electornic": ["wirelees earbuds", "usb-c hub","phone stad"],
    "Clothing": ["Running shoes","Hoodie","Denum Jacket"],
    "Home and Kichen": ["Air fyer","coffee maker", "Blender"],
    "Sports": ["Yoga Mat","Water Bottle","Resistance Bands"],
    "Book": ["Python cookbook", "Data Engineering Book "],
    "Beauty": ["Face Serum","Sunscreen SPF50","Lip Balm"],
    "Toys": ["LEGO Set","Board Game","Puzzle 1000pc"]
}

customers = [{"customer_id":i,"name":fake.name(),
  "email":fake.email(),"city":fake.city(),
  "country":fake.country_code(),
  "signup_date":fake.date_between(start_date="-2y",end_date="today")}
  for i in range(1,501)]

df_customers = pd.DataFrame(customers)
df_customers.to_csv("customers.csv", index=False)

products = []
for i in range(1,51):
    cat=random.choice(CATEGORES)
    products.append({"product_id":i,
      "product_name":random.choice(PRODUCT_NAME[cat])+f" #{i}",
      "category":cat,
      "base_price":round(random.uniform(5.99,299.99),2)})
df_products = pd.DataFrame(products)
df_products.to_csv("products.csv", index=False)

orders = []
for i in range(1,5001):
    pid=random.randint(1,50)
    price=df_products.loc[df_products.product_id==pid,"base_price"].values[0]
    qty=random.randint(1,5)
    orders.append({"order_id":i,"customer_id":random.randint(1,500),
      "product_id":pid,"quantity":qty,"unit_price":price,
      "total_price":round(price*qty,2),
      "order_date":fake.date_between(start_date="-1y",end_date="today"),
      "status":random.choice(STATUSES)})
df_orders = pd.DataFrame(orders)
df_orders.to_csv("orders.csv", index=False)

print("=== Validation ===")
print(f"Customers : {len(df_customers):,} rows | nulls: {df_customers.isnull().sum().sum()}")
print(f"Products  : {len(df_products):,} rows  | nulls: {df_products.isnull().sum().sum()}")
print(f"Orders    : {len(df_orders):,} rows  | nulls: {df_orders.isnull().sum().sum()}")
print(f"Status split:\n{df_orders.status.value_counts()}")
print("\nFiles saved: customers.csv, products.csv, orders.csv")
```
