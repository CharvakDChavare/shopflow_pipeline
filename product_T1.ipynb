```python
from faker import Faker
import pandas as pd
import random
```

```python
CATEGORES = [
    "Electornic", "Clothing", "Home and Kichen", "Sports", "Book", "Beauty", "Toys"
]

PRODUCT_NAME={
    "Electornic": ["wirelees earbuds", "usb-c hub","phone stad"],
    "Clothing": ["Running shoes","Hoodie","Denum Jacket"],
    "Home and Kichen": ["Air fyer","coffee maker", "Blender"],
    "Sports": ["Yoga Mat","Water Bottle","Resistance Bands"],
    "Book": ["Python cookbook", "Data Engineering Book "],
    "Beauty": ["Face Serum","Sunscreen SPF50","Lip Balm"],
    "Toys": ["LEGO Set","Board Game","Puzzle 1000pc"]
}

products = []
for i in range(1, 51):
    cat = random.choice(CATEGORES)
    name_list = PRODUCT_NAME[cat]
    products.append({
        "product_id": i,
        "product_name": random.choice(name_list) + f"#(i)",
        "categres": cat,
        "base_price": round(random.uniform(5.99, 299.99), 2)
    })

    df_products = pd.DataFrame(products)
    df_products.to_csv("products.csv", index=False)
    print(f"Generated {len(df_products)} products ")
```
