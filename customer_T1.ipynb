```python
from faker import Faker
import pandas as pd
import random

fake = Faker()
random.seed()
Faker.seed()
```

```python
customers = []

for i in range(1, 501):
    customers.append({
        "customer_id": i,
        "name": fake.name(),
        "email": fake.email(),
        "city": fake.city(),
        "country": fake.country_code(),
        "singup_data": fake.date_between(
            start_date="-2y",
            end_date="today"
        )
    })
    
    df_coustomers = pd.DataFrame(customers)
    df_coustomers.to_csv("customers.csv", index=False)
    print(f"Generated {len(df_coustomers)} customers")
```
